```python
df = pd.read_csv("weather dataset.csv")

print("First 5 rows of dataset:")
print(df.head())
```

```python
print("Checking Missing Values:")
print(df.isnull().sum())

df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
print("Dataset Information:")
print(df.info())

print("Statistical Summary:")
print(df.describe())
```

```python
plt.figure()
plt.plot(df["temp_mean(c)"])

plt.title("Temperature Trend Over Time")
plt.xlabel("Record Index")
plt.ylabel("Temperature (°C)")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['global_radiation'], bins=30, edgecolor='black')
plt.title('Distribution of Global Radiation')
plt.xlabel('Global Radiation')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
print("Descriptive Statistics for Global Radiation:")
print(df['global_radiation'].describe())
```

```python
plt.figure(figsize=(10, 6))
plt.boxplot(df['global_radiation'])
plt.title('Box Plot of Global Radiation')
plt.ylabel('Global Radiation')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
plt.figure()
plt.plot(df["temp_mean(c)"])

plt.title("Temperature Trend Over Time")
plt.xlabel("Record Index")
plt.ylabel("Temperature (°C)")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Using available columns: 'normalized_label' instead of 'Season' and 'temp_max(c)' instead of 'Rainfall'
label_temp = df.groupby('normalized_label')['temp_max(c)'].mean()

plt.figure(figsize=(10, 6))
label_temp.plot(kind='bar', color='skyblue', edgecolor='black')

plt.title('Average Max Temperature by Normalized Label')
plt.xlabel('Normalized Label')
plt.ylabel('Average Max Temp (°C)')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure()
plt.hist(df["temp_mean(c)"], bins=10)

plt.title("Temperature Distribution")
plt.xlabel("Temperature")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure()
plt.scatter(df["temp_mean(c)"], df["global_radiation"])

plt.title("Temperature vs Global Radiation")
plt.xlabel("Temperature")
plt.ylabel("Global Radiation")
plt.show()
```
